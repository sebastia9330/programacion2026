LLM NODELS (GPT3.5)

```python
from langchain_openai import OpenAI
```

```python
gpt3 = OpenAI(model_name='gpt-3.5-turbo-instruct', temperature=0.7, max_tokens=1024)
```

```python
print(gpt3)
```

```python
respuesta = gpt3.invoke("Explica qué es un modelo de IA LLM")
```

```python
print(respuesta)
```

```python
respuestas = gpt3.generate(['Cual es la capital de colombia',
                            'que es oracle',
                            'se puede conectar un llm a oracle'])
```

```python
print(respuestas)
```

```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI
```

```python
chat = ChatOpenAI(model_name='gpt-3.5-turbo', temperature=0.5, max_tokens=1024)
```

```python
mensajes = [
    SystemMessage(content = 'Eres un experto en ingenieria de datos y quieres ser mi maestro'),
    HumanMessage(content = 'Escribe los pasos mas importantes para ser el mejor ingeniero de datos del mundo y ser un referente en la materia')
]
respuesta = chat.invoke(mensajes)
```
