```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("palabras.xlsx")
df = df.dropna().reset_index(drop=True)
df
```

```python
palabras = list(df['palabras'])
palabras
```

```python
def get_embedding(text, model="text-embedding-3-small"):
    response = cliente.embeddings.create(
        model=model,
        input=text
    )
    return response.data[0].embedding
```

```python
df = df[df['palabras'].apply(lambda x: isinstance(x, str))]
df['embedding'] = df['palabras'].apply(get_embedding)

df
```

```python
df.to_csv('palabrasEmbbedigs.csv', sep=';', index=False)
```

```python
searchTerm = 'abeja'
searchTermVector = get_embedding(searchTerm)
print(searchTermVector)
```

```python
def cosine_similarity(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
df['similitud'] = df['embedding'].apply(
    lambda x: cosine_similarity(x, searchTermVector)
)

df = df.sort_values('similitud', ascending=False)
df
```
